```python
import spacy
import pandas as pd
from collections import Counter
```

```python
nlp = spacy.load("en_core_web_sm")

df = pd.read_excel('../data/Questions.xlsx')
```

```python
for index, row in df.iterrows(): 
    question = nlp(row['Question'])
    tags = [token.lemma_ for token in question]
    print(row['Type'],'-----', row['Question'],'-----', 'NNS' in tags or 'VBP' in tags)
```

```python
doc1 = nlp("Apples are ")
doc2 = nlp("The apple is yummy")


for token in doc1:
    print(token.text, token.lemma_, token.pos_, token.tag_, spacy.explain(token.tag_))
```
